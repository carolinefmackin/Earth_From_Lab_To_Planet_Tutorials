```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy.stats import linregress
excel_file_path = 'elasticity_lab_blank_for_students3.xlsx'
excel_data = pd.read_excel(excel_file_path)
#print(excel_data)
excel_data.columns = ['Band','Mass (g)','Mass (kg)','F (N)','A (m2)','F/A (Pa)','X0  (cm)','X0  (m)','xf (cm)','xf (m)','∆x (m)','L (m)','∆x/L','Unload (m)','E (Pa)','k']
row_indices_1 = [1,2,3,4,5]
selected_data_1 = pd.to_numeric(excel_data.loc[row_indices_1, '∆x/L'], errors='coerce')
selected_data_2 = pd.to_numeric(excel_data.loc[row_indices_1, 'F/A (Pa)'], errors='coerce')
row_indices_2 = [10,11,12,13,14]
selected_data_3 = pd.to_numeric(excel_data.loc[row_indices_2, '∆x/L'], errors='coerce')
selected_data_4 = pd.to_numeric(excel_data.loc[row_indices_2, 'F/A (Pa)'], errors='coerce')
row_indices_3 = [19,20,21,22,23]
selected_data_5 = pd.to_numeric(excel_data.loc[row_indices_3, '∆x/L'], errors='coerce')
selected_data_6 = pd.to_numeric(excel_data.loc[row_indices_3, 'F/A (Pa)'], errors='coerce')
print(selected_data_1)
print(selected_data_2)
print(selected_data_3)
print(selected_data_4)
print(selected_data_5)
print(selected_data_6)
```

```python
plt.scatter(selected_data_1, selected_data_2, label='Data Set 1')
slope_1, intercept_1, r_value, p_value, std_err = linregress(selected_data_1, selected_data_2)
plt.plot(selected_data_1, slope_1 * selected_data_1 + intercept_1, color='red', label='Linear Regression Set 1')
plt.scatter(selected_data_3, selected_data_4, label='Data Set 2')
slope_2, intercept_2, r_value, p_value, std_err = linregress(selected_data_3, selected_data_4)
plt.plot(selected_data_3, slope_2 * selected_data_3 + intercept_2, color='blue', label='Linear Regression Set 2')
plt.scatter(selected_data_5, selected_data_6, label='Data Set 3')
slope_3, intercept_3,r_value, p_value, std_err = linregress(selected_data_5, selected_data_6)
plt.plot(selected_data_5, slope_3 * selected_data_5 + intercept_3, color='green', label='Linear Regression Set 3')
plt.xlabel('∆x/L')
plt.ylabel('F/A (Pa)')
plt.title('Scatter Plot with Linear Regression')
plt.legend()
```
