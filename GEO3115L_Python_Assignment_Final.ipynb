# <span style="font-size: 24px; font-weight: bold;">Part 4: Adding Linear Regression, Error Bars, and Further Analysis</span>

# <span style="font-size: 18px; font-weight: bold;">4.1: Find Error Bars for the first 8 months in 2023</span>




```python
# I added these data points for simplicity. Normally you would upload these from the csv 
data = [63.5, 67.3, 74.5, 76.5, 84.2, 94.8, 99.8, 103.4]

# Step 1: Calculate the mean
n = len(data)  # len(data) = number of data points
mean = sum(data) / n
print(f"Mean: {mean:.2f}")

# Step 2: Calculate each deviation from the mean and square it
squared_deviations = [(x - mean) ** 2 for x in data]

# Step 3: Calculate the variance
# the average of the squared differences between each data point in a set and the mean of that set
variance = sum(squared_deviations) / (n - 1)  

# Step 4: Calculate the standard deviation
std_dev = variance ** 0.5  # Square root of variance
print(f"Standard Deviation (Manual): {std_dev:.2f}")

# Step 5: Calculate the standard error of the mean (SEM)
sem = std_dev / ((n) ** 0.5)
print(f"Standard Error of the Mean (Manual): {sem:.2f}")
```

# <span style="font-size: 18px; font-weight: bold;">The Short Cut</span>. . . 

```python
# Import numpy
import numpy as np

# Calculate mean, standard deviation, and SEM using numpy
mean_np = np.mean(data)
std_dev_np = np.std(data, ddof=1)  # Use ddof=1 for sample standard deviation
sem_np = std_dev_np / (np.sqrt(len(data)))

# Print results
print(f"Mean (numpy): {mean_np:.2f}")
print(f"Standard Deviation (numpy): {std_dev_np:.2f}")
print(f"Standard Error of the Mean (numpy): {sem_np:.2f}")
```

Summary of Key Concepts:

- Sample Mean:
The sample mean is an estimate of the true mean of the entire population.

- Standard Deviation (SD):
The standard deviation measures the variability within a dataset by indicating how far individual data points typically deviate from the mean. A higher SD reflects greater spread in the data.

- Standard Error of the Mean (SEM):
SEM measures how accurate the sample mean is as an estimate of the population mean. A smaller SEM suggests a more reliable mean and less variability in repeated sampling.


# <span style="font-size: 18px; font-weight: bold;">4.3: Fit a linear Regression</span>

Let's look at the relationship between the months January - August and average temperature for those months
- Fit a linear regression for "Month" (January - August only) vs. "2023 Average Temp"
- plot the data points and the regression line 

```python
from scipy.stats import linregress

#data just January - August 
#always update this to your specifc file location
data = pd.read_csv("Austin_Temperature_Data.csv")
data_jan_to_aug = data[:8]
temps = data_jan_to_aug['2023 Average Temp']

n = len(temps)
mean_temp = temps.mean()
std_dev = temps.std()  # Standard Deviation
sem = std_dev / (n ** 0.5)  # Standard Error of the Mean

# Perform linear regression (Month index as x, temperatures as y)
x = np.arange(1, n + 1)  # Month indices 1 to 8
y = temps.values
slope, intercept, r_value, p_value, std_err = linregress(x, y)

print(f"Mean Temp: {mean_temp:.2f}")
print(f"Standard Deviation (STD): {std_dev:.2f}")
print(f"Standard Error of the Mean (SEM): {sem:.2f}")
print(f"Linear Regression Slope: {slope:.2f}, Intercept: {intercept:.2f}")
print(f"R-squared: {r_value**2:.2f}")
```

```python
# Step 1: Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Step 2: Load the data (USE YOUR FILE PATH)
data = pd.read_csv("Austin_Temperature_Data.csv")

# Step 3: Filter data for January to August
data_jan_to_aug = data[:8]  # Select the first 8 rows (Jan-Aug)
temps = data_jan_to_aug['2023 Average Temp']  

# Step 4: Calculate stats
n = len(temps)  # Number of data points
mean_temp = temps.mean()  # Mean temperature
std_dev = temps.std(ddof=1)  # Standard deviation
sem = std_dev / (n ** 0.5)  # Standard Error of the Mean (SEM)

# Step 5: linear regression
x = np.arange(1, n + 1)  # Month indices (1 to 8)
y = temps.values  # Temperature values
#r_value is the correlation coefficient 
#p_value is the statistical significance. 
slope, intercept, r_value, p_value, std_err = linregress(x, y)  # Linear regression


# Step 6: Format the regression equation for the legend
regression_eq = f"y = {slope:.2f}x + {intercept:.2f}"

# Step 7: Create the plot
plt.figure(figsize=(10, 6))  # Set figure size

# Step 8: Scatter plot with error bars
plt.errorbar(
    x, y, yerr=std_dev, fmt='o', color='red', ecolor='red', 
    elinewidth=1, capsize=3, label='2023 Average Temp (+/- 1 SD)'
)

# Step 9: Plot
plt.plot(
    x, intercept + slope * x, color='red', linestyle='--', 
    label=f'Linear Regression: {regression_eq}'
)

# Step 10: Customize 
plt.xlabel("Month", fontsize=12)  # X-axis label
plt.ylabel("Temperature (°F)", fontsize=12)  # Y-axis label
plt.title("Average Monthly Temperatures in 2023 (Jan-Aug)", fontsize=14)  # Title
plt.xticks(
    ticks=x, labels=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August'], 
    rotation=45  # Rotate month labels for readability
)
plt.legend(fontsize=10)  # Add legend
plt.grid(True)  # Add grid for better visualization

# Step 11: Show the plot
plt.show()

# Step 12: Calculate and print RMS
rms = np.sqrt(np.mean(y**2))  # Root Mean Square
print(f"Root Mean Square (RMS) of the temperatures: {rms:.2f} °F")
```

## Residuals

Is the straight line a good fit? One can always, and should, eyeball the fit by looking at the match of the line to the data. However, it is easier to see when we plot the *residual*, that is the difference between the "model" (here, a straight line) and the data. Let's try that.


```python
# this is the function we think the data should follow, we got it from linear regression
model = intercept + slope * x
# let's plot the difference from the model
plt.scatter(x,y-model,label='residual');plt.grid(True)
# make this symmetrical so that it is easier to spot bias
plt.ylim(-5,5)
plt.xlabel("month number", fontsize=12)
plt.ylabel("Temperature residual (°F)", fontsize=12)
plt.show()
```

# <span style="font-size: 24px; font-weight: bold;">Part 5: Plotting Multiple Independent Variables (years in this case)</span>


## <span style="font-size: 18px; font-weight: bold;">5.1: Multiple Lines on One Plot</span>


```python
# Import necessary packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load the data source
data = pd.read_csv("Austin_Temperature_Data.csv")

# Create a function to calculate and plot the linear regression
def plot_linear_regression(data, year, color, marker, linestyle):
    # Filter January to August data
    temps = data[:8][f'{year} Average Temp']
    
    # Define x and y
    x = np.arange(1, len(temps) + 1)  # Month indices 1 to 8
    y = temps.values
    
    # Calculate the coefficients for the regression line
    coefficients = np.polyfit(x, y, 1)
    regression_line = np.poly1d(coefficients)
    
    # Calculate STD
    std_dev = np.std(y, ddof=1)  # Sample standard deviation
    std_value = f"STD = {std_dev: .2f}"
    
    # Plot data with error bars (using STD)
    plt.errorbar(
        x, y, yerr=std_dev, fmt=marker, color=color, ecolor=color,
        elinewidth=1, capsize=3, label=f'{year} Average Temp ({std_value})'
    )
    
    # Scatter plot
    plt.scatter(x, y, color=color, label=f'{year} Data', marker=marker)
    
    # Plot the regression line
    plt.plot(x, regression_line(x), color=color, linestyle=linestyle, label=f'{year} Regression Line')
    
    #calculate RMS 
    rms = np.sqrt(np.mean(y**2))  # Root Mean Square
    print(f"Root Mean Square (RMS) of temperatures for {year}: {rms:.2f} °F")

# Plotting setup
plt.figure(figsize=(12, 8))

# Add data for 2000, 2010, and 2023
plot_linear_regression(data, 2000, 'green', 'o', '--')
plot_linear_regression(data, 2010, 'orange', 'o', '--')
plot_linear_regression(data, 2023, 'red', 'o', '--')

# Add labels, title, legend, and grid
plt.xlabel("Month", fontsize=12)
plt.ylabel("Temperature (°F)", fontsize=12)
plt.title("Average Monthly Temperatures (Jan-Aug) for 2000, 2010, and 2023", fontsize=14)
plt.xticks(ticks=np.arange(1, 9), labels=['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August'], rotation=45)
plt.legend(fontsize=10)
plt.grid(True)

# Show the plot
plt.show()
```

## <span style="font-size: 18px; font-weight: bold;">Task 5.2: Let's find the R^2 values for each year</span>

```python
# Function to calculate R^2 for a given year
def calculate_r_squared(data, year):
    # January to August data
    temps = data[:8][f'{year} Average Temp']
    
    x = np.arange(1, len(temps) + 1)  # Month indices 1 to 8
    y = temps.values
    # _ is used here to ignore certain outputs we do not need in this specific calculation
    _, _, r_value, _, _ = linregress(x, y)
    
    return r_value ** 2


years = [2000, 2010, 2023]

# Calculate and print R^2 values
for year in years:
    r_squared = calculate_r_squared(data, year)
    print(f"R² for {year}: {r_squared:.2f}")
```

The \( R^2 \) value, or coefficient of determination, measures how well a linear regression model fits the data. Specifically:
- It represents the proportion of the variance in the dependent variable (temperature) that is predictable from the independent variable (month).
- An \( R^2 \) value close to 1 indicates that the model explains most of the variation in the data.
- An \( R^2 \) value closer to 0 suggests the model does not fit well.
