```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import linregress
data= pd.read_csv("Book1.csv")
print(data)

x=data['one'];
y=data['two'];
plt.scatter(x,y)
slope_1, intercept_1, r_value, p_value, std_err = linregress(x, y)
plt.plot(x, slope_1 * x + intercept_1, color='red', label='Linear Regression Set 1')

#These line of code are new and we diid not do in class 
#This is how you add labels, a title, and a legend
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.legend()
plt.title('Add Title')

#Code to add the equation of linear regression to plot 
equation_text = f'Equation: y = {slope_1:.2f}x + {intercept_1:.2f}'
plt.text(2, 5.5, equation_text, fontsize=10, color='blue')
```
